# 판다스 문제 풀어보기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

```python
# insurance.csv'파일 읽어서 df에 할당하고, 상위 5개 항목 출력하기

df = pd.read_csv('abc/data/insurance.csv', encoding="utf-8")
df.head()
```

```python
# df의 요약(평균, 4분위수, 전체 갯수 등)정보 확인하기
df.describe()
```

```python
#신규 컬럼(double bmi)을 생성하고, bmi의 2배가 되는 값을 할당하기
df["double_bmi"] = df["bmi"] * 2
df.head()
```

```python
# 신규 컬럼(debt)을 생성하고, 0으로 채우기
df['debt'] = 0
df.head()
```

```python
# debt에 0부터 값이 증가하도록 채우기
df['debt'] = range(len(df))
df.head()
```

```python
# 1번 인덱스의 정보 출력
df.iloc[1]
```

```python
# 'double bmi'열 삭제하기
df = df.drop(["double_bmi"], axis=1)
df.head()
```

```python
# 'age'열, 'sex'열만 출력하기
df[["age", "sex"]].head()
```

```python
# 나이를 1살씩 증가시키기
df["age"] = df["age"]+1
df.head()
```

```python
# charges의 표시형식을 소수점 아래 2자리로 바꾸기
# df["charges"].apply(lambda x: round(x, 2))
df["charges"] = df["charges"].astype(float)
df["charges"] = df["charges"].map(lambda x: round(x,2))
df.head()
```

```python
# 컬럼 정렬(알파벳 순서대로) : age, bmi, charges... 순으로 정렬
df = df.sort_index(axis=1)
df.head()
```

```python
# 값 정렬(age) : 나이 순으로 정렬
df = df.sort_values(by="age", ascending=True)
df
```

```python
# 나이('age') 평균구하기
df_am = df["age"].mean()
df_am
```

```python
# 남자, 여자가 각각 몇명있는지 구하기
df["sex"].value_counts()
```

```python
# 자녀(children)가 3명인 데이터 출력하기
df[df["children"] == 3]
```

```python
# 자녀(children)가 5명 이상인 데이터 출력하기
df[df["children"] >= 5]
```

```python
# bmi가 17 미만인 데이터 출력하기
df[df["bmi"] < 17]
```

```python
# 성별에 따른 보험료 평균
df.groupby("sex")["charges"].mean().round()
```

```python
# 흡연여부에 따른 보험료 평균
df.groupby("smoker")["charges"].mean().round()
```

```python
# 지역별 보험료 평균
df.groupby("region")["charges"].mean().round()
```

```python
# 흡연여부 + 지역별 데이터 평균
df_m = df.groupby(["smoker" , "region"])[["age", "bmi", "charges", "children", "debt"]].mean().round()
df_m
```

```python
# 흡연여부 + 지역별 데이터 평균에서 'charges' 만 츨력하기
df_m["charges"]
```

```python
# 흡연여부 + 지역별 데이터 크기(size)
df.groupby(["smoker", "region"]).size()
```

```python
# smoker 여부에 따라, group_val 컬럼에 '흡연자', '비흡연자' 표시하기
df["group_val"] = np.where(df["smoker"] == "yes", "흡연자", "비흡연자")
df.head()
```

```python
df = df.sort_index(axis=1)
df.head()
```
