```python
# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print(df)

```

```python
def age_category(age):
    
    if age < 30:
        return '20대'  

    elif age < 40:
        return '30대'
    else:
        return '40대 이상'
    
age_category(25)
```

```python
# Category 컬럼으로 추가
df['Age_Category'] = df['Age'].apply(age_category)
df
```

```python
# 보너스
def calc_bonus(salary):
    return salary*0.30

df['Bonus'] = df['Salary'].apply(lambda x: calc_bonus(x))
df
```

## MERGE

```python
#매출 데이터
sales_data = pd.DataFrame({
    'CustomerID': [101, 102, 103, 104],
    'Date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-07'],
    'Amount': [250, 300, 150, 400]
})

#고객 데이터
customer_data = pd.DataFrame({
    'CustomerID': [101, 102, 103, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Eva'],
    'Region': ['North', 'South', 'East', 'West']
})
#매출 데이터
sales_data = pd.DataFrame({
    'CustomerID': [101, 102, 103, 104],
    'Date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-07'],
    'Amount': [250, 300, 150, 400]
})

#고객 데이터
customer_data = pd.DataFrame({
    'CustomerID': [101, 102, 103, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Eva'],
    'Region': ['North', 'South', 'East', 'West']
})
```

```python
sales_data
```

```python
customer_data
```

```python
customer_data
```

```python
pd.merge(sales_data, customer_data, on='CustomerID') #조건 기준으로 같은 것만 나오는 inner와 같다
```

```python
pd.merge(sales_data, customer_data, on='CustomerID', how='outer')
```

```python
#매출기준
pd.merge(sales_data, customer_data, on='CustomerID', how='left')
```

```python
#고객기준
pd.merge(sales_data, customer_data, on='CustomerID', how='right')
```
