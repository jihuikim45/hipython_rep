```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6))
plt.title('5일간 주식가격')
plt.xlabel('Date')
plt.ylabel('Price')
plt.ylim(95, 110)
plt.grid()

plt.plot(dates, prices, label='Stock Price')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2, 1)
plt.plot(dates, prices)
plt.xticks(rotation=40)

plt.subplot(1,2, 2)
plt.bar(dates, prices)
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart.png')
```

```python
data = (np.random.randn(50)+1).cumsum()
plt.plot(data)
```

```python
# 도화지 만들기: 1행 3열짜리 subplot
plt.figure(figsize=(15, 4))  # 가로 15, 세로 4 크기의 그림

# 1️⃣ 선그래프
plt.subplot(1, 3, 1)  # 1행 3열 중 첫 번째
plt.plot(data)
plt.title('Line Plot')

# 2️⃣ 막대그래프
plt.subplot(1, 3, 2)
plt.bar(range(len(data)), data)
plt.title('Bar Chart')

# 3️⃣ 점그래프 (산점도)
plt.subplot(1, 3, 3)
plt.scatter(range(len(data)), data)
plt.title('Scatter Plot')

# 그래프 전체 보여주기
plt.tight_layout()
plt.show()
```
