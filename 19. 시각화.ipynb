```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
tips_df["day"]
```

```python
tips_df.describe()
```

```python
# 1번량 그래프: 히스토그램

sns.histplot(data=tips_df, x="total_bill", bins=20, edgecolor=None)
```

```python
# 2분할 도화지 만들기
fig = plt.figure(figsize=(12,5))
axes1 = fig.add_subplot(1,2,1) #total bill
axes2 = fig.add_subplot(1,2,2) #tips

axes1.hist(data=tips_df, x="total_bill", bins=20)
axes2.hist(data=tips_df, x="tip", bins=20)
plt.tight_layout()
plt.show()
```

```python
# boxplot

fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(data=tips_df, x="total_bill")
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes =fig.add_subplot(1,1,1)
axes.boxplot(
    
    x=[
        tips_df[tips_df['sex']=='Female']['total_bill'],
        tips_df[tips_df['sex']=='Male']['total_bill']          
    ],
    labels=['Female', 'Male']
)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(data=tips_df, x="size")
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes =fig.add_subplot(1,1,1)
axes.boxplot(
    
    x=[    
        tips_df[tips_df['day']=='Thur']['total_bill'],
        tips_df[tips_df['day']=='Fri']['total_bill'],
        tips_df[tips_df['day']=='Sat']['total_bill'],
        tips_df[tips_df['day']=='Sun']['total_bill']
                 
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun']
)
plt.show()
```

```python
fig = plt.figure(figsize=(16,5))

# 1. 시간별 매출
axes1 = fig.add_subplot(1, 3, 1)
axes1.boxplot(
    x=[
        tips_df[tips_df['time'] == 'Dinner']['total_bill'],
        tips_df[tips_df['time'] == 'Lunch']['total_bill']
    ],
    labels=['Dinner', 'Lunch']
)
axes1.set_title('시간별 매출')

# 2. 요일별 매출
axes2 = fig.add_subplot(1, 3, 2)
axes2.boxplot(
    x=[
        tips_df[tips_df['day'] == 'Thur']['total_bill'],
        tips_df[tips_df['day'] == 'Fri']['total_bill'],
        tips_df[tips_df['day'] == 'Sat']['total_bill'],
        tips_df[tips_df['day'] == 'Sun']['total_bill']
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun']
)
axes2.set_title('요일별 매출')

# 3. 요일별 팁
axes3 = fig.add_subplot(1, 3, 3)
axes3.boxplot(
    x=[
        tips_df[tips_df['day'] == 'Thur']['tip'],
        tips_df[tips_df['day'] == 'Fri']['tip'],
        tips_df[tips_df['day'] == 'Sat']['tip'],
        tips_df[tips_df['day'] == 'Sun']['tip']
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun']
)
axes3.set_title('요일별 팁')

plt.tight_layout()
plt.show()
```

```python
# scatter 산점도
fig = plt.figure(figsize=(7,5))
axes =fig.add_subplot(1,1,1)
axes.scatter(tips_df['total_bill'], tips_df['tip'])
plt.show()
```

```python
colors = {"Female":"tomato", "Male":"blue"}


# 다변량 표현
# 성별에 따라 색상 주기 (C속성)
# 값의 크기에 따라 점의 크기 설정 (S속성)

# scatter 산점도
fig = plt.figure(figsize=(7,5))
axes =fig.add_subplot(1,1,1)
axes.scatter(
tips_df['total_bill'],tips_df['tip'],
s=tips_df['size']**2*10, #tablesize
c=tips_df['sex'].map(colors)
)
plt.show()
```

```python
colors = {"Female":"#CD5555", "Male":"#483D8B"} #16진수 색상코드

tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']
fig = plt.figure()
axes =fig.add_subplot(1,1,1)
axes.scatter(
tips_df['total_bill'],tips_df['tip'],
s=tips_df['tip_pct']*1000, #tablesize, tip
c=tips_df['sex'].map(colors),
alpha=0.5
)
plt.show()
```

# 다변량 표현 


# 파이 차트 : 비율


```python
# 파이 차트 : 비율
#성별비율비교
#빈도수



fig = plt.figure()
sex_counts = tips_df['sex'].value_counts()

axes =fig.add_subplot(1,1,1)
axes.pie(
sex_counts,
colors=['blue','red'],
labels=sex_counts.index,
autopct='%1.1f%%',
startangle=90
)
plt.title(성별)
plt.show()
```

```python


fig = plt.figure(figsize=(5,5))

# 1. 성별
sex_counts = tips_df['sex'].value_counts()
axes1 = fig.add_subplot(2, 2, 1)
axes1.pie(
    sex_counts,
    colors=['#87CEFA', '#FFC0CB'],
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
axes1.set_title('성별')

# 2. 흡연 여부
smoker_counts = tips_df['smoker'].value_counts()
axes2 = fig.add_subplot(2, 2, 2)
axes2.pie(
    smoker_counts,
    colors=['#87CEFA', '#FFC0CB'],
    labels=smoker_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
axes2.set_title('흡연 여부')

# 3. 요일별 손님 비율
day_counts = tips_df['day'].value_counts()
axes3 = fig.add_subplot(2, 2, 3)
axes3.pie(
    day_counts,
    colors=['#87CEFA', '#FFC0CB', '#FF9999', '#FFCC99'],
    labels=day_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
axes3.set_title('요일별 손님 비율')

# 4. 일행 수
size_counts = tips_df['size'].value_counts()
axes4 = fig.add_subplot(2, 2, 4)

axes4.pie(
    size_counts,
    colors = ['#87CEFA', '#FFC0CB', '#FF9999', '#FFCC99', '#C299FF', '#FFB266'],    
    labels=size_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
axes4.set_title('일행 수 비율')

plt.tight_layout()
plt.show()
```
