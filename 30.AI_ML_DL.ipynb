# 머신러닝

규칙이 아닌 모델을 선정
모델 : 데이터를 넣어준다 > 학습    
 fit()  
모델 : 값을 넣어 결과를 확인한다 > 예측

```python
import numpy as np
from sklearn import linear_model
```

```python
X_train = numpy.array([1,2,3,4,5,6,7]).reshape(-1,1)
X_train
```

```python
y_train = np.array([12,17,22,27,32,37,42])
y_train
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict([[9]])
```

# 딥러닝
모델이 정해져 있다.   
계층설정: 입력값, 은닉층, 출력값   
모델 생성(계층을 넣어서)   
모델 설정: compile   
모델 학습: 데이터를 넣어서 (fit)   
학습한 모델로 예측: predict


```python
import tensorflow as tf
```

```python
X_train = numpy.array([1,2,3,4,5,6,7]).reshape(-1,1)
y_train = np.array([12,17,22,27,32,37,42])
```

```python
inputs = tf.keras.layers.Input(shape=(1,))
outputs = tf.keras.layers.Dense(units=1)(inputs)
```

```python
model = tf.keras.models.Model(inputs, outputs)
```

```python
# 모델 설정
model.compile(
    loss = tf.keras.losses.MeanSquaredError(),
    optimizer = tf.keras.optimizers.SGD()
)
```

```python
# 데이터 넣고 학습시키기
model.fit(X_train, y_train, epochs=10000, verbose=0)
```

```python
# 예측
model.predict(np.array([[9]]))
```
