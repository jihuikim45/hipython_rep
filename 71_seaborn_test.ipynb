# boxplot
이상치 탐지, 통계량 확인, 분포 확인   

```python
ev_df = pd.read_csv('data1/EV_charge.csv')
ev_df.head()
```

```python
ev_df.info()
```

```python
sns.boxplot(
    x='weekday', y='kwhTotal', data=ev_df
)
```

원본데이터 + 통계량을 같이 표현한다 > boxplot + stripplot    

boxlpot이 놓치는 것이 많기 때문에 만능은 아니다.


```python
sns.boxplot(
    x='weekday', y='kwhTotal', data=ev_df
)

sns.stripplot(
    x='weekday', y='kwhTotal', data=ev_df
)
```

# swarmplot
- 값들이 겹치지 않는다.   
- 분포를 잘 보여준다.  
- 성능 저하를 고려해야 한다. 

```python
sns.boxplot(
    x='weekday', y='kwhTotal', data=ev_df
)

sns.stripplot(
    x='weekday', y='kwhTotal', data=ev_df
)

sns.swarmplot(
    x='weekday', y='kwhTotal', data=ev_df, color='grey', alpha=0.4
)
```

```python
weekday_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sns.boxplot(x='weekday', y='kwhTotal', data=ev_df, hue='platform', order=weekday_order)
```

```python
weekday_order = {'weekday':[]}
```

```python
#plotly boxplot


px.box(
    data_frame=ev_df, x='weekday', y='kwhTotal',
    width=500, height=500, points='all', category_orders={'weekday' : weekday_order},
    color='platform'
).show()
```

E heatmap
테이블 요약(합계, 평균, ---)한 결과를 색으로 시각화     
- 요약: 피봇 테이블(탐색)   
- 인사이트 시각화 히트맵     



```python
# 요약
pivot_df = tip_df.pivot_table('tip', 'day', 'size')
#시각화
sns.heatmap(pivot_df, annot=True)
```

```python
pivot_df.fillna(0, inplace=True)
```
